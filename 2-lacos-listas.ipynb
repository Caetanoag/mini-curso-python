````python
import builtins

def testar(solucao, casos):
    original_input = builtins.input
    original_print = builtins.print
    passou = 0

    for i, (entrada, esperado) in enumerate(casos):
        linhas = iter(entrada.split('\n'))
        saida_capturada = []

        builtins.input = lambda _='': next(linhas)
        builtins.print = lambda *args, sep=' ', end='\n', **kw: saida_capturada.append(sep.join(str(a) for a in args) + end)

        try:
            solucao()
        except StopIteration:
            builtins.input = original_input
            builtins.print = original_print
            original_print(f"💥 Caso {i+1}: ERRO — solução chamou input() mais vezes que linhas na entrada")
            continue
        except Exception as e:
            builtins.input = original_input
            builtins.print = original_print
            original_print(f"💥 Caso {i+1}: ERRO — {type(e).__name__}: {e}")
            continue
        finally:
            builtins.input = original_input
            builtins.print = original_print

        resultado = ''.join(saida_capturada).strip()
        if resultado == esperado.strip():
            original_print(f"✅ Caso {i+1}: PASSOU")
            passou += 1
        else:
            original_print(f"❌ Caso {i+1}: FALHOU")
            original_print(f"   Entrada : {entrada!r}")
            original_print(f"   Esperado: {esperado.strip()!r}")
            original_print(f"   Obtido  : {resultado!r}")

    original_print(f"\n{'='*30}\n   {passou}/{len(casos)} casos passaram\n{'='*30}")

print('✅ Setup concluído! Pode começar.')
````

## 🔁 Bloco 1 — Laço `for` e `range`

O `for` repete um bloco de código para cada elemento de uma sequência.
O `range` gera sequências de números sem ocupar memória.

| Forma | O que gera |
|-------|------------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 |
| `range(5, 0, -1)` | 5, 4, 3, 2, 1 |

````python
# Contando de 1 a 5
for i in range(1, 6):
    print(i)
````

````python
# Padrão acumulador — soma de 1 a N
n = int(input())  # digita um número, ex: 10

soma = 0
for i in range(1, n + 1):
    soma += i

print(soma)
````

````python
# Padrão contagem — quantos pares de 1 a N
n = int(input())  # digita um número, ex: 10

pares = 0
for i in range(1, n + 1):
    if i % 2 == 0:
        pares += 1

print(f"{pares} pares de 1 até {n}")
````

````python
# Padrão muito comum na OBI: ler N valores um por linha
n = int(input())  # digita 3
for i in range(n):
    x = int(input())  # digita um número por vez
    print(f"Li: {x}")
````

## 🔄 Bloco 2 — Laço `while`

O `while` repete enquanto uma condição for verdadeira. Usar quando **não sabemos de antemão quantas vezes repetir**.

| Situação | Use |
|----------|-----|
| Quantidade de repetições conhecida | `for` |
| Repetir até uma condição mudar | `while` |
| Ler até um valor de parada (`0`, `-1`, etc.) | `while` |

````python
# Contagem regressiva
n = int(input())  # digita 5

while n > 0:
    print(n)
    n -= 1

print("Fogo!")
````

````python
# Padrão sentinela: ler até o usuário digitar 0
soma = 0
x = int(input())  # digita números; 0 encerra

while x != 0:
    soma += x
    x = int(input())

print(soma)
````

````python
# break e continue
i = 0
while True:         # loop infinito...
    i += 1
    if i % 2 == 0:
        continue    # pula pares
    if i > 10:
        break       # ...até aqui
    print(i)
````

## 📋 Bloco 3 — Listas

Lista é uma sequência de valores indexada a partir de **0**.

| Operação | Código | Resultado |
|----------|--------|-----------|
| Criar | `lst = [3, 1, 4, 1, 5]` | — |
| Acessar | `lst[0]` | `3` |
| Último | `lst[-1]` | `5` |
| Tamanho | `len(lst)` | `5` |
| Adicionar | `lst.append(9)` | — |
| Ordenar | `lst.sort()` | — |
| Mínimo/Máximo | `min(lst)` / `max(lst)` | — |
| Soma | `sum(lst)` | — |

````python
# Criando e acessando
notas = [7.5, 9.0, 6.0, 8.5, 5.0]

print(notas[0])      # primeira nota
print(notas[-1])     # última nota
print(len(notas))    # quantidade
print(sum(notas))    # soma
print(max(notas))    # maior
print(min(notas))    # menor
````

````python
# Lendo uma lista da entrada — padrão OBI
n = int(input())                           # digita 4
valores = list(map(int, input().split()))  # digita: 10 3 7 1

print(f"Lista:  {valores}")
print(f"Sorted: {sorted(valores)}")  # sorted() não modifica a original
print(f"Lista após sorted: {valores}")
valores.sort()                        # sort() modifica no lugar
print(f"Lista após sort1: {valores}")
````

````python
# Iterando sobre lista
nums = [4, 2, 7, 1, 9]

# Jeito 1: pelo valor
for x in nums:
    print(x, end=' ')
print()

# Jeito 2: pelo índice
for i in range(len(nums)):
    print(f"nums[{i}] = {nums[i]}")
````

## 🔤 Bloco 4 — Strings

String é uma sequência de caracteres. Em Python, strings se comportam como listas de letras.

| Operação | Código | Resultado |
|----------|--------|-----------|
| Tamanho | `len("oi")` | `2` |
| Acesso | `"abc"[0]` | `'a'` |
| Fatiamento | `"abcde"[1:3]` | `'bc'` |
| Maiúsculo | `"oi".upper()` | `'OI'` |
| Minúsculo | `"OI".lower()` | `'oi'` |
| Separar | `"a b c".split()` | `['a','b','c']` |
| Juntar | `" ".join(['a','b'])` | `'a b'` |
| Pertence | `'a' in "abc"` | `True` |

````python
# Iterando sobre string — igual a lista
palavra = input()  # digita uma palavra, ex: python

for letra in palavra:
    print(letra, end='-')
print()
````

````python
# Contando vogais
palavra = input().lower()  # digita uma palavra
vogais = 'aeiou'

count = 0
for c in palavra:
    if c in vogais:
        count += 1

print(count)
````
