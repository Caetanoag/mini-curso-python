````python
import builtins

def testar(solucao, casos):
    original_input = builtins.input
    original_print = builtins.print
    passou = 0

    for i, (entrada, esperado) in enumerate(casos):
        linhas = iter(entrada.split('\n'))
        saida_capturada = []

        # Intercepta input() e print() da solução do aluno
        builtins.input = lambda _='': next(linhas)
        builtins.print = lambda *args, sep=' ', end='\n', **kw: saida_capturada.append(sep.join(str(a) for a in args) + end)

        try:
            solucao()
        except StopIteration:
            builtins.input = original_input
            builtins.print = original_print
            original_print(f"💥 Caso {i+1}: ERRO — solução chamou input() mais vezes que linhas na entrada")
            continue
        except Exception as e:
            builtins.input = original_input
            builtins.print = original_print
            original_print(f"💥 Caso {i+1}: ERRO — {type(e).__name__}: {e}")
            continue
        finally:
            builtins.input = original_input
            builtins.print = original_print

        resultado = ''.join(saida_capturada).strip()
        if resultado == esperado.strip():
            original_print(f"✅ Caso {i+1}: PASSOU")
            passou += 1
        else:
            original_print(f"❌ Caso {i+1}: FALHOU")
            original_print(f"   Entrada : {entrada!r}")
            original_print(f"   Esperado: {esperado.strip()!r}")
            original_print(f"   Obtido  : {resultado!r}")

    original_print(f"\n{'='*30}\n   {passou}/{len(casos)} casos passaram\n{'='*30}")

print('✅ Setup concluído! Pode começar.')
````

## 📋 Bloco 1 — O que é a OBI Fase 1?

A **OBI (Olimpíada Brasileira de Informática)** tem duas modalidades:
- **Iniciação:** algoritmos mais simples, foco em lógica
- **Programação:** algoritmos mais elaborados

Na **Fase 1**, os problemas são resolvidos **submetendo código** a um juiz online automático. Seu programa:

1. **Lê a entrada** do teclado (stdin)
2. **Processa** os dados
3. **Imprime a resposta** exata na tela (stdout)

### ⚠️ Regras de ouro da OBI:
| Regra | Por quê importa |
|-------|-----------------|
| A saída deve ser **exatamente** igual ao esperado | Um espaço a mais = resposta errada |
| Nada de `input("Digite: ")` | A mensagem vai para o stdout e quebra o juiz |
| Leia **todos** os dados antes de processar | A entrada segue um formato fixo descrito no enunciado |
| Não use `print()` extra para debug | Eles aparecem na saída e geram WA (Wrong Answer) |

### Veredictos do juiz:
- ✅ **AC** (Accepted) — correto!
- ❌ **WA** (Wrong Answer) — resposta errada
- ⏱️ **TLE** (Time Limit Exceeded) — código muito lento
- 💥 **RE** (Runtime Error) — erro de execução (ex: divisão por zero)

### Tipos de dados e casting

| Tipo | Exemplo | Quando usar |
|------|---------|-------------|
| `int` | `42` | Números inteiros |
| `float` | `3.14` | Números decimais |
| `str` | `"oi"` | Texto |
| `bool` | `True` / `False` | Verdadeiro/Falso |

**Casting** = converter de um tipo para outro.

````python
# input() sempre retorna string — isso aqui DÁ ERRADO:
a = input()   # digita 5
b = input()   # digita 3
print(a + b)  # o que acontece?
````

### Formatando a saída

A saída precisa ser **exata**. Aprenda as formas de formatar:

````python
x = 42
pi = 3.14159

# f-string (recomendado)
print(f"O valor é {x}")

# Separador personalizado
print(1, 2, 3, sep=" ")   # 1 2 3
print(1, 2, 3, sep="\n")  # um por linha

# Sem pular linha no final
print("Sem newline", end=" ")
print("continua aqui")

# Casas decimais (quando o problema pede)
print(f"{pi:.2f}")   # 3.14
print(f"{pi:.4f}")   # 3.1416
````

## ➗ Bloco 3 — Operadores lógicos e aritméticos

| Operador | Nome | Exemplo | Resultado |
|----------|------|---------|-----------|
| `+` `-` `*` | Básicos | `7 + 3` | `10` |
| `/` | Divisão real | `7 / 2` | `3.5` |
| `//` | Divisão inteira | `7 // 2` | `3` |
| `%` | Resto (módulo) | `7 % 2` | `1` |
| `**` | Potência | `2 ** 10` | `1024` |


````python
# Verificar se número é par ou ímpar
n = 17
if n % 2 == 0:
    print("par")
else:
    print("ímpar")
````

````python
# Extrair dígitos de um número
n = 1234
unidade  = n % 10         # 4
dezena   = (n // 10) % 10 # 3
centena  = (n // 100) % 10# 2
milhar   = n // 1000      # 1

print(f"Milhar:{milhar} Centena:{centena} Dezena:{dezena} Unidade:{unidade}")
````

````python
# Converter segundos em horas/minutos/segundos
total = 3723  # segundos

horas   = total // 3600
minutos = (total % 3600) // 60
segundos = total % 60

print(f"{horas}h {minutos}min {segundos}s")
````

 ### 🔵 Problema A — Aprovado ou Reprovado


---



No Colégio Neps Academy (CNA) um aluno é aprovado por média se ele obtiver uma média final maior ou igual a 7, caso o aluno tenha uma média menor que 7 mas maior ou igual a 4 ele está de recuperação, caso ele tenha uma média menor que 4 o aluno está reprovado.

A média é calculada com a nota das duas provas aplicadas no semestre e corresponde simplesmente a média aritimética das duas notas.

Baseado nas duas notas do aluno, indique o resultado final do aluno: "Aprovado", "Reprovado" ou "Recuperacao".


---


**Entrada:** A entrada consiste de apenas uma linha com as notas
A e B, que correspondem as duas notas que o aluno conquistou esse semestre.

**Saìda esperada:** A saída do seu programa deve ser apenas uma linha. Caso o aluno tenha sido aprovado informe "Aprovado", caso o aluno tenha sido reprovado informe "Reprovado" e caso ele esteja de recuperação informe "Recuperacao".


---


#### **Exemplos**
**Entrada:** `4.3 5`  
**Saída:** Recuperacao

**Entrada:** `10 9.8`  
**Saída:** Aprovado

---



````python
def solucao():
  nota1, nota2 = map(float, input().split())
  media = (nota1 + nota2) / 2

  if media >= 7:
    print("Aprovado")
  elif media >= 4:
    print("Recuperacao")
  else:
    print("Reprovado")
````
