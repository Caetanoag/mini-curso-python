````python
import builtins

def testar(solucao, casos):
    original_input = builtins.input
    original_print = builtins.print
    passou = 0

    for i, (entrada, esperado) in enumerate(casos):
        linhas = iter(entrada.split('\n'))
        saida_capturada = []

        # Intercepta input() e print() da soluĂ§ĂŁo do aluno
        builtins.input = lambda _='': next(linhas)
        builtins.print = lambda *args, sep=' ', end='\n', **kw: saida_capturada.append(sep.join(str(a) for a in args) + end)

        try:
            solucao()
        except StopIteration:
            builtins.input = original_input
            builtins.print = original_print
            original_print(f"đ„ Caso {i+1}: ERRO â soluĂ§ĂŁo chamou input() mais vezes que linhas na entrada")
            continue
        except Exception as e:
            builtins.input = original_input
            builtins.print = original_print
            original_print(f"đ„ Caso {i+1}: ERRO â {type(e).__name__}: {e}")
            continue
        finally:
            builtins.input = original_input
            builtins.print = original_print

        resultado = ''.join(saida_capturada).strip()
        if resultado == esperado.strip():
            original_print(f"â Caso {i+1}: PASSOU")
            passou += 1
        else:
            original_print(f"â Caso {i+1}: FALHOU")
            original_print(f"   Entrada : {entrada!r}")
            original_print(f"   Esperado: {esperado.strip()!r}")
            original_print(f"   Obtido  : {resultado!r}")

    original_print(f"\n{'='*30}\n   {passou}/{len(casos)} casos passaram\n{'='*30}")

print('â Setup concluĂ­do! Pode comeĂ§ar.')
````

## đ Bloco 1 â O que Ă© a OBI Fase 1?

A **OBI (OlimpĂ­ada Brasileira de InformĂĄtica)** tem duas modalidades:
- **IniciaĂ§ĂŁo:** algoritmos mais simples, foco em lĂłgica
- **ProgramaĂ§ĂŁo:** algoritmos mais elaborados

Na **Fase 1**, os problemas sĂŁo resolvidos **submetendo cĂłdigo** a um juiz online automĂĄtico. Seu programa:

1. **LĂȘ a entrada** do teclado (stdin)
2. **Processa** os dados
3. **Imprime a resposta** exata na tela (stdout)

### â ïž Regras de ouro da OBI:
| Regra | Por quĂȘ importa |
|-------|-----------------|
| A saĂ­da deve ser **exatamente** igual ao esperado | Um espaĂ§o a mais = resposta errada |
| Nada de `input("Digite: ")` | A mensagem vai para o stdout e quebra o juiz |
| Leia **todos** os dados antes de processar | A entrada segue um formato fixo descrito no enunciado |
| NĂŁo use `print()` extra para debug | Eles aparecem na saĂ­da e geram WA (Wrong Answer) |

### Veredictos do juiz:
- â **AC** (Accepted) â correto!
- â **WA** (Wrong Answer) â resposta errada
- â±ïž **TLE** (Time Limit Exceeded) â cĂłdigo muito lento
- đ„ **RE** (Runtime Error) â erro de execuĂ§ĂŁo (ex: divisĂŁo por zero)

### Formatando a saĂ­da

A saĂ­da precisa ser **exata**. Aprenda as formas de formatar:

````python
x = 42
pi = 3.14159

# f-string (recomendado)
print(f"O valor Ă© {x}")

# Separador personalizado
print(1, 2, 3, sep=" ")   # 1 2 3
print(1, 2, 3, sep="\n")  # um por linha

# Sem pular linha no final
print("Sem newline", end=" ")
print("continua aqui")

# Casas decimais (quando o problema pede)
print(f"{pi:.2f}")   # 3.14
print(f"{pi:.4f}")   # 3.1416
````

## â Bloco 3 â Operadores lĂłgicos e aritmĂ©ticos

| Operador | Nome | Exemplo | Resultado |
|----------|------|---------|-----------|
| `+` `-` `*` | BĂĄsicos | `7 + 3` | `10` |
| `/` | DivisĂŁo real | `7 / 2` | `3.5` |
| `//` | DivisĂŁo inteira | `7 // 2` | `3` |
| `%` | Resto (mĂłdulo) | `7 % 2` | `1` |
| `**` | PotĂȘncia | `2 ** 10` | `1024` |


````python
# Verificar se nĂșmero Ă© par ou Ă­mpar
n = 17
if n % 2 == 0:
    print("par")
else:
    print("Ă­mpar")
````

````python
# Extrair dĂ­gitos de um nĂșmero
n = 1234
unidade  = n % 10         # 4
dezena   = (n // 10) % 10 # 3
centena  = (n // 100) % 10# 2
milhar   = n // 1000      # 1

print(f"Milhar:{milhar} Centena:{centena} Dezena:{dezena} Unidade:{unidade}")
````

````python
# Converter segundos em horas/minutos/segundos
total = 3723  # segundos

horas   = total // 3600
minutos = (total % 3600) // 60
segundos = total % 60

print(f"{horas}h {minutos}min {segundos}s")
````
