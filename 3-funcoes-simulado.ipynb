````python
import builtins

def testar(solucao, casos):
    original_input = builtins.input
    original_print = builtins.print
    passou = 0

    for i, (entrada, esperado) in enumerate(casos):
        linhas = iter(entrada.split('\n'))
        saida_capturada = []

        builtins.input = lambda _='': next(linhas)
        builtins.print = lambda *args, sep=' ', end='\n', **kw: saida_capturada.append(sep.join(str(a) for a in args) + end)

        try:
            solucao()
        except StopIteration:
            builtins.input = original_input
            builtins.print = original_print
            original_print(f"💥 Caso {i+1}: ERRO — solução chamou input() mais vezes que linhas na entrada")
            continue
        except Exception as e:
            builtins.input = original_input
            builtins.print = original_print
            original_print(f"💥 Caso {i+1}: ERRO — {type(e).__name__}: {e}")
            continue
        finally:
            builtins.input = original_input
            builtins.print = original_print

        resultado = ''.join(saida_capturada).strip()
        if resultado == esperado.strip():
            original_print(f"✅ Caso {i+1}: PASSOU")
            passou += 1
        else:
            original_print(f"❌ Caso {i+1}: FALHOU")
            original_print(f"   Entrada : {entrada!r}")
            original_print(f"   Esperado: {esperado.strip()!r}")
            original_print(f"   Obtido  : {resultado!r}")

    original_print(f"\n{'='*30}\n   {passou}/{len(casos)} casos passaram\n{'='*30}")

print('✅ Setup concluído! Pode começar.')
````

## 🔧 Bloco 1 — Funções

Funções organizam o código em blocos reutilizáveis. Na OBI, usar funções evita repetição e deixa a solução mais fácil de depurar.

```python
def nome(parametros):
    # corpo
    return resultado
```

| Conceito | Detalhe |
|----------|---------|
| `def` | Define a função |
| Parâmetros | Valores de entrada (opcionais) |
| `return` | Devolve um valor (sem `return` → devolve `None`) |
| Escopo | Variáveis dentro da função não existem fora dela |

````python
# Função simples que retorna o quadrado da entrada
def quadrado(n):
    return n * n

print(quadrado(5))   # 25
print(quadrado(10))  # 100
````

````python
# Funções que retornam True/False são muito úteis na OBI
def e_par(n):
    return n % 2 == 0

def e_divisivel(n, d):
    return n % d == 0

print(e_par(4))           # True
print(e_par(7))           # False
print(e_divisivel(12, 3)) # True
````

````python
# Verificar se número é primo
# Um número é primo se só é divisível por 1 e por ele mesmo
def e_primo(n):
    if n < 2:
        return False
    for i in range(2, n):      # testa todos os divisores de 2 até n-1
        if n % i == 0:
            return False       # achou divisor → não é primo
    return True

for n in [1, 2, 3, 4, 13, 15, 17]:
    print(f"{n}: {'primo' if e_primo(n) else 'não primo'}")
````

## 🧩 Bloco 2 — Padrões Clássicos da OBI

Alguns padrões aparecem repetidamente na OBI. Reconhecê-los de imediato economiza muito tempo na prova.

#### Encontre o valor máximo e minimo dentro de uma lista

````python
# Padrão: máximo e mínimo manual (sem usar max/min)
# Útil quando você precisa rastrear o índice ou comparar condicionalmente

nums = [3, 7, 1, 9, 4, 2]

maior = nums[0]   # começa assumindo que o primeiro é o maior
menor = nums[0]

for x in nums:
    if x > maior:
        maior = x
    if x < menor:
        menor = x

print(f"Maior: {maior}, Menor: {menor}")
````

#### Encontre a posição de um elemento dado

````python
# Padrão: busca linear — encontrar posição de um elemento
nums = [5, 3, 8, 1, 9, 2]
alvo = int(input())   # digita o número que quer encontrar

encontrado = -1       # -1 = não encontrado
for i in range(len(nums)):
    if nums[i] == alvo:
        encontrado = i
        break

if encontrado != -1:
    print(f"Encontrado na posição {encontrado}")
else:
    print("Não encontrado")
````

#### Diga se uma lista tem valores negativos ou não

````python
# Padrão: flag booleana — verificar se ALGUM elemento satisfaz condição
nums = list(map(int, input().split()))  # digita: 4 7 2 9 1

tem_negativo = False
for x in nums:
    if x < 0:
        tem_negativo = True
        break

print("sim" if tem_negativo else "nao")
````

Dado um valor l, representando o indice de inicio de um intervalo, e r representando o final, mostre a soma de todos elementos dentro da lista no intervalo list[l] até list[r]

````python
# Padrão: soma de prefixo — soma acumulada até cada posição
# Muito comum em problemas de subarray e intervalos
nums = [1, 2, 3, 4, 5]

prefixo = [0] * (len(nums) + 1) # Gera um array preenchido com zeros e com tamanho len(nums) + 1
for i in range(len(nums)):
    prefixo[i + 1] = prefixo[i] + nums[i]
print(prefixo)  # [0, 1, 3, 6, 10, 15]

l, r = 1, 3   # soma de nums[1] até nums[3]
print(prefixo[r + 1] - prefixo[l])  # 2+3+4 = 9
````

## 📖 Bloco 3 — Dicionários

Dicionário guarda pares **chave → valor**. É a estrutura certa quando precisamos associar dados ou contar frequências.

| Operação | Código | Resultado |
|----------|--------|-----------|
| Criar | `d = {}` | — |
| Inserir | `d['a'] = 1` | — |
| Acessar | `d['a']` | `1` |
| Chave existe? | `'a' in d` | `True` |
| Default seguro | `d.get('x', 0)` | `0` se não existir |
| Iterar | `for k, v in d.items()` | — |
| Chaves | `d.keys()` | — |
| Valores | `d.values()` | — |

````python
# Criando e acessando
aluno = {'nome': 'Ana', 'nota': 9.5, 'turma': 'A'}

print(aluno['nome'])           # Ana
print(aluno.get('idade', 0))   # 0 — chave não existe, retorna padrão

aluno['nota'] = 10.0           # atualiza
aluno['cidade'] = 'SP'         # insere

for chave, valor in aluno.items():
    print(f"{chave}: {valor}")
````

````python
# Padrão mais importante: contador de frequência
texto = input()  # digita: banana

freq = {}
for c in texto:
    freq[c] = freq.get(c, 0) + 1

for letra, qtd in freq.items():
    print(f"'{letra}': {qtd}")
````
